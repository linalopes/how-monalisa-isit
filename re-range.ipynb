```python
# Import necessary libraries
from skimage import io, color
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the image
# Replace 'image_path.jpg' with the path to your image file
image = io.imread('data/woody_bulle.JPG')
```

```python
# Step 2: Separate the image into R, G, B channels
# Ensure the image has three channels
red_channel = image[:, :, 0]
green_channel = image[:, :, 1]
blue_channel = image[:, :, 2]
```

```python
# Step 3: Flatten each channel and sort its values
sorted_red = np.sort(red_channel.flatten())  # Sort the red channel
sorted_green = np.sort(green_channel.flatten())  # Sort the green channel
sorted_blue = np.sort(blue_channel.flatten())  # Sort the blue channel
```

```python
# Step 4: Reshape the sorted channels back to the original image shape
sorted_red_image = sorted_red.reshape(red_channel.shape)
sorted_green_image = sorted_green.reshape(green_channel.shape)
sorted_blue_image = sorted_blue.reshape(blue_channel.shape)
```

```python
# Step 5: Stack the sorted channels back into an RGB image
sorted_image = np.stack((sorted_red_image, sorted_green_image, sorted_blue_image), axis=2)
```

```python
# Step 6: Display the original and the reorganized image
plt.figure(figsize=(12, 6))
```

```python
# Original image
plt.subplot(1, 2, 1)
plt.title("Original Image", fontsize=14)
plt.imshow(image)
plt.axis('off')

# Reorganized image (Sorted by R, G, B channels)
plt.subplot(1, 2, 2)
plt.title("Reorganized Image (R, G, B Sorted)", fontsize=14)
plt.imshow(sorted_image.astype(np.uint8))  # Ensure the image is displayed as uint8
plt.axis('off')

# Define the resolution for the output
output_dpi = 300  # Higher DPI for better quality
output_file = "comparison_high_res.png"  # Name of the saved file

# Save the figure with high resolution
plt.tight_layout()
plt.savefig(output_file, dpi=output_dpi, bbox_inches='tight')

plt.show()
print(f"Comparison image saved as '{output_file}' in high resolution.")
```
